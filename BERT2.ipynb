```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.initializers import TruncatedNormal
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.metrics import CategoricalAccuracy
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Input, Dense, Lambda

# Define input layers using tf.keras.Input
input_ids = Input(shape=(max_len,), dtype=tf.int32, name="input_ids")
input_mask = Input(shape=(max_len,), dtype=tf.int32, name="attention_mask")

# Wrap the BERT call in a Lambda layer and specify output_shape
embeddings = Lambda(lambda x: bert(x[0], attention_mask=x[1])[0],
                    output_shape=(max_len, bert.config.hidden_size))([input_ids, input_mask])
# output_shape is set to (max_len, hidden_size of BERT)

out = tf.keras.layers.GlobalMaxPool1D()(embeddings)
out = Dense(128, activation='relu')(out)
out = tf.keras.layers.Dropout(0.1)(out)
out = Dense(32, activation='relu')(out)

y = Dense(2, activation='softmax')(out)

# Create the model using KerasTensors as inputs
model = tf.keras.Model(inputs=[input_ids, input_mask], outputs=y)
model.layers[2].trainable = True  # Adjust the layer index if needed
```

```python
optimizer = Adam(
    learning_rate=5e-05, # HF recommendation
    epsilon=1e-08,
    decay=0.01,
    clipnorm=1.0
)

loss = CategoricalCrossentropy(from_logits=True)
metric = CategoricalAccuracy('balanced_accuracy')

model.compile(
    optimizer=optimizer,
    loss=loss,
    metrics=[metric]
)
```

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Assuming `train_labels` is your array of training labels
class_weights = compute_class_weight(class_weight='balanced',
                                     classes=np.array([0, 1]),
                                     y= df_train['label'])

class_weights_dict = {0: class_weights[0], 1: class_weights[1]}
print("Class Weights:", class_weights_dict)
```

```python
history = model.fit(
    x = {'input_ids':X_train['input_ids'], 'attention_mask':X_train['attention_mask']},
    y = to_categorical(df_train['label']),
    validation_data = ({'input_ids':X_test['input_ids'], 'attention_mask':X_test['attention_mask']},
                        to_categorical(df_test['label'])),
    epochs=3,
    batch_size=32,class_weight=class_weights_dict
)
```

```python
from sklearn.metrics import classification_report

predicted = model.predict({'input_ids': X_test['input_ids'], 'attention_mask': X_test['attention_mask']})
y_predicted = np.argmax(predicted, axis=1)
print(classification_report(df_test['label'], y_predicted))
```
